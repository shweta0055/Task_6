# Lets split up the data into a training set and a test set

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop('Species',axis=1)
#contains the target
y=data['Species']
```

```python
x.shape
```

```python
y.shape
```

# Split the data into train and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print('Shape of Traing data :',x_train.shape)
print('Shape of test data: ',x_test.shape)
```

# Decision Tree Model Training

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(x_train, y_train)
```

```python
predictions =dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,10))
tree=plot_tree(dtree,feature_names=data.columns,precision=2,rounded=True,filled=True,class_names=y.values)
```
